```python
#Slice the data
Y1=data1.iloc[:,0]
X1=data1.iloc[:,2:32]
X1
```

```python
# Train-Test Split for Model Evaluation
X_train, X_test, Y_train, Y_test = train_test_split(X1, Y1, test_size=0.2, random_state=42, shuffle=True)
```

```python
X_train_2d = X_train.values.reshape(-1, 1)
X_test_2d = X_test.values.reshape(-1, 1)
Y_train_2d = Y_train.values.reshape(-1, 1)
Y_test_2d = Y_test.values.reshape(-1, 1)
```

```python
data2.set_index("ID", inplace=True)
```

```python
data2
```

```python
#Slice the data

data2cleaned=data2.iloc[:,1:31]
data2cleaned
```

```python
# Initialize and train all six models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "Perceptron": Perceptron(),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(n_neighbors=5),  # You can adjust the number of neighbors
}


model_accuracies = {}

for model_name, model in models.items():
    model.fit(X_train, Y_train)  # Train the model
    Y_pred = model.predict(X_test)  # Make predictions
    accuracy = accuracy_score(Y_test, Y_pred)  # Calculate accuracy
    model_accuracies[model_name] = accuracy

# Print the evaluation results on Data1.csv
for model_name, accuracy in model_accuracies.items():
    print(f"{model_name} Accuracy on Data1.csv:", accuracy)


# Predict Y for Data2 using all six models
predictions_data2 = {}

for model_name, model in models.items():
    Y_pred_data2 = model.predict(data2cleaned)
    predictions_data2[model_name] = Y_pred_data2

# Create a DataFrame containing the predictions
predictions_data2_df = pd.DataFrame(predictions_data2)

# Save the predictions as a CSV file
predictions_data2_df.to_csv('Data2_Predictions.csv', index=True)
```
